## Question:-1

     Consider the Iris ML data set (Download it from University of Californiaâ€“Irvine Machine Learning Data Repository). Generate the variance covariance matrix (Matrix Dimension should be 4 cross 4, since the data has only 4 attributes). Then find out the first two principal components of it.

```python
import pandas as pd
import numpy as np
from sklearn.datasets import load_iris
from sklearn.preprocessing import StandardScaler
```

```python
iris = load_iris()
iris_data = iris.data
```

```python
iris_data
```

```python
scaler = StandardScaler()
iris_data_scaled = scaler.fit_transform(iris_data)
```

```python
cov_matrix = np.cov(iris_data_scaled.T)
```

```python
cov_matrix
```

```python
eigenvalues, eigenvectors = np.linalg.eig(cov_matrix)
```

```python
eigenvalues
```

```python
eigenvectors
```

```python
idx = eigenvalues.argsort()[::-1]
eigenvalues = eigenvalues[idx]
eigenvectors = eigenvectors[:, idx]
```

```python
first_two_pc = eigenvectors[:, :2]
```

```python
print("Variance-Covariance Matrix:")
print(cov_matrix)
print("\nFirst two principal components:")
print(first_two_pc)
```
